# 객체에 대한 반복처리

- iterable 객체(str, list, tuple, set, dict)
- range 객체

```python
name = '홍길동'
for ch in name:
  print(ch)

fruits = ['🍏','🍑','🍉','🥑','🍅']
for fruit in fruits:
  print(fruit)
```

```python
# range 객체
# - range(start, end, step)
# - 반복횟수와 범위

for i in range(0, 10, 3):
  print(i)
print()

for n in range(2, 101, 2):
  print(n, end=' ')

print()

for n in range(1, 51):
  print(n * 2, end=' ')
```

```python
# 1~100까지 홀수의 합은?
total = 0

for n in range(1, 101, 2):
  #print(n, end=' ')
  total += n

print(f'1~100사이 홀수의 합은 {total} 입니다.')
```

```python
# 2~9단까지 한번에 출력
for dan in range(2, 10):
  print('=' * 5, dan, '단', '=' * 5)

  for n in range(1, 10):
    print(f'{dan} * {n} = {dan * n}')

    print()
```
