# 함수 function

- 절차적인, 반복적인 실행코드를 가진 자료형.
- 선언 후에 호출해서 사용한다.
- 호출시에 함수부로 값을 전달할 수 있고, 함수 실행결과를 반환 받을 수 있다.

```python
def foo(a, b):
  print(f'{a=}, {b=}')
  return a + b

print(foo(10, 20))
print(type(foo))
print(foo)
display(foo)
```

# 매개변수 parameter | 매개인자(인수) argument

- 매개변수: 함수선언부의 변수(공간)
- 매개인자: 함수호출부에서 함수부로 전달되는 값. 매개인자가 매개변수에 대입되며 함수가 실행된다.

```python
# 함수선언부
def bar(k, m): # 매개변수
  print(f'{k= }, {m= }')

# 함수 호출부
bar(3, 5) # 3, 5 > 매개 인자

a = 4
b = 7
bar(a, b) # 매개인자
```

# 리턴

- 함수선언부에서 함수실행을 마치고, 다시 함수선언부로 돌아오는 리턴
- 리턴시에 함께 값이 돌아올 수 있다.

```python
def zoo():
  print('zoo called!')
  return 100

print('-' * 10)
result = zoo()
print(result)
print('*' * 10)
```

# 호출방식

1. 위치 매개인자 전달방식: 전달한 매개인자가 선언된 매개변수에 순서대로 대입
2. 키워드 매개인자 전달방식: 전달 순서와 상관없이 동일한 name의 매개변수에 대입

```python
def test(a, b):
    print(f'{a= }, {b= }')

test(10, 20)   # 위치 매개인자 전달방식
test(a=1, b=2) # 키워드 매개인자 전달방식
test(b=2, a=1) 
test(30, b=40)

# # test(40, a=50)
# test(a=60, 50)
# test(a=1, b=2, c=3) # 존재하지 않는건 넣을 수 없음
# test(10)
```

# 기본값 지정

```python
# 기본값 지정한 매개변수는 함수 호출 시 값을 전달하지 않아도 된다(선택)

def print_info(name, age, has_married=True):
  print(f'{name}은/는 {age}세이고, 결혼 {"했습니다" if has_married else "하지 않았습니다."}')

print_info('홍길동', 33, True)
print_info('신사임당', 22) # defualt가 True
# print_info('신사임당') # 기본값이 안된 매개변수는 생략할 수 없다.
```

# packing 연산자
- 여러개의 매개인자를 하나의 (매개)변수에 대입할 수 있다.
- *args: n개의 위치 매개인자를 받음
- **kwargs: n개의 키워드 매개인자를 받음

```python
def accept(*args):
  print(args, type(args))

accept()
accept(1)
accept(1, 2, 3)
```

```python
# packing 매개변수는 일반매개변수의 오른쪽에 선언해야한다.

# def hello(*scores, name):
def hello(name, *scores):
  print(f'{name}의 점수: {scores}')

hello('전봉준', 100, 30, 50)
hello('신사임당', 90, 80)
```

```python
def spec(**a):
  print(a, type(a))

spec()
spec(width=300, height=700, color=['white', 'red', 'brown'])
spec(width=300, height=700)
```

```python
def user_info(name, age, **configs):
  print(f'{name = }')
  print(f'{age = }')
  print(f'{configs = }')

user_info('홍길동', 22, hobbys=['넷플릭스', '축구'], addr= '서울시 도봉구')
```

```python
def sample(a, b, c=99, *args, **kwargs):
  print(a, b, c, args, kwargs)

sample(10, 20, 30, 40, 50, weather='rainy', temperature=12)
```

# unpacking 연산자

- 컨테이너객체의 요소를 개별적으로 전달하는 방법
- 연산자는 packing연산자와 동일하지만, 사용되는 위치에 따라 구분할 수 있다.
- 변수부가 아닌 값을 사용하는 부분에 unpacking 연산자를 사용할 수 있다.

```python
def test(*args): # 함수선언부 - packing 연산자
  print(args)

names = ['홍길동', '홍진경', '홍난파']

test(names)
test(names[0], names[1], names[2])

test(*names) # 함수호출부 - unpacking 연산자
test('홍길동', '홍진경', '홍난파')
```

# 1급시민객체 함수

- 함수는 값으로써 처리될 수 있는 1급시민객체이다.
- 변수에 대입할 수 있다.
- 리턴값으로써 사용될 수 있다.

```python
def foo():
  print('foooooooooooooooo')

print(foo, type(foo))

doo = foo
print(doo, type(doo))
```

```python
def runner(f, n=10):
  """함수실행기"""
  for i in range(n):
    f()
# runner(foo)
runner(foo, 3)
```
