# 입출력 IO 

## open 함수

`open()` 함수는 파이썬에서 **파일을 열기 위한 기본 함수**이다. 파일을 읽거나 쓰기 위해 사용되며, 다양한 모드로 파일을 열 수 있다.


### 기본 문법

```python
f = open('파일경로', '모드', encoding='인코딩')
```

| 인자       | 설명                                      |
|------------|-------------------------------------------|
| 파일경로   | 열고자 하는 파일의 경로 또는 이름          |
| 모드       | 파일을 여는 방법 (읽기, 쓰기 등)           |
| encoding   | 문자 인코딩 (텍스트 파일인 경우 지정 필요) |

---

### 주요 모드 

| 모드 | 의미                    | 파일 없을 때         |
|------|-------------------------|------------------------|
| `'r'`  | 읽기 모드                | 오류 발생 (`FileNotFoundError`) |
| `'w'`  | 쓰기 모드 (덮어쓰기)     | 새로 생성됨             |
| `'a'`  | 추가 모드 (이어쓰기)     | 새로 생성됨             |
| `'x'`  | 배타적 생성 모드         | 이미 존재하면 오류       |
| `'b'`  | 바이너리 모드 (추가)     | 이진 데이터 처리용       |
| `'t'`  | 텍스트 모드 (기본)       | 생략 가능 (`'rt'` = `'r'`) |

> 예: `'rb'` (이진 읽기), `'wt'` (텍스트 쓰기), `'a'` (텍스트 추가)


### 추가 팁

- `read()` : 전체 내용 문자열로 읽기
- `readline()` : 한 줄씩 읽기
- `readlines()` : 모든 줄을 리스트로 반환
- `write()` : 문자열 1개 작성
- `writelines()` : 리스트 형식 문자열 여러 개 작성



## 파일 쓰기

````python
text = """안녕하세요~ 파이썬으로 만든 귀도입니다.
I Love you~ 
1234567890
🍇🥝🍋‍🟩🍌
"""

f = open('sample.txt', 'w', encoding='utf-8')
print(f, type(f))

f.write(text)
f.close()  # 반드시 객체를 반환해야 메모리 누수가 생기지 않는다.
````

# 파일 읽기

````python
f = open('sample.txt', 'r', encoding='utf-8')

print(f.read())

f.close()
````

````python
# with블럭: 생성된 객체를 자동으로 반환
with open('sample.txt', 'w', encoding='utf-8') as f:
    f.write(text)

with open('sample.txt', 'r', encoding='utf-8') as f:
    print(f.read())
````

````python
# for문을 통해 한줄씩 읽기
with open('sample.txt', 'r', encoding='utf-8') as f:
    for line in f:
        print(line, end='')
````

````python
# readline(): 한번에 한줄씩 읽어 반환
with open('sample.txt', 'r', encoding='utf-8') as f:
    # := 바다코끼리 연산자: 대입연산자 + 좌항값반환
    while line := f.readline():   
        print(line, end='')
````

````python
# 이어쓰기
with open('sample.txt', 'a', encoding='utf-8') as f:
    f.write('Thank you 😊😊😊😊😊\n')    
````
