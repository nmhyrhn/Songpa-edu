# 형변환 casting

## 컴퓨터 연산 원칙

  1. 동일한 자료형 사이에서만 연산이 가능하다.
  2. 연산 결과 역시 동일한 자료형이다.
  
  - int + float : float + float 형변환 후 연산처리
  - int + bool : int + int 형변환 후 연산처리

# 암묵적 형변환

- 자동으로 형변환 후 연산되는 경우

```python
print(3 + 1.23) # 3.0 + 1.23
print(3 + True) # 3 + 1
# print(3 + "5") 
print(3 + int("5")) 
```

```python
# bool 타입으로 암묵적 형변환
# - 값이 존재하는 것: True
# - 값이 없는 것(0, '', [], {}): False
if '':
  print('🍩🍩🍩') # 출력 안됨
if 'a':
  print('🍩🍩🍩') # 출력 가능
```

# 명시적 형변환

```python
x = int(input('점수입력: '))
print(type(x))
print(x - 1)
print(float('3.5') + 4)
```

```python
# 컨테이너 자료형 간 변환
# list(), tuple(), set(), dict()

print(list((1, 2, 3))) # tuple -> list
print(tuple([1, 2, 3])) # list -> tuple
print(set([1, 2, 2, 3]))# list - > set
print(set((1, 2, 2, 3))) # tuple -> set
print(dict([('a', 100), ('b', 200)])) # list 안에 tuple -> dict
print(list({'a': 100, 'b': 200})) # dict -> list / key 값만 정리됨
```
