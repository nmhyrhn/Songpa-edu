# tuple

시퀀스객체이면서 불변(immutable)의 객체 순서를 저장

```python
t1 = () # 빈 튜플을 생성
print(t1, type(t1))

t2 = (10,)
print(t2, type(t2))

# t3 = (10, 20)
t3 = 10, 20

print(t3, type(t3))
```

```python
tpl = 1, 2, 3
print(tpl)
print(tpl[0])
print(tpl[0:2])

print(tpl * 2)
print(tpl)

# tpl[0] = 100 불가함
#tpl[0:2] = (100, 200)
```

```python
hong = ('홍길동', 33, [100, 89, 55])

# hong[0] = '신사임당'
hong[2][2] = 70 # 주소값 안에 있는 값이 변경된거므로 변경 가능함
```

```python
# 값 치환
x = 33
y = 55

# temp = x
# x = y
# y = temp

x, y = y, x # python에서는 가능함

print(f'x= {x}, y= {y}')
```

```python
foods = ('🍩', '🎂', '🍬', '🍭', '🎂')

for food in foods:
  print(food)

# enumerate() 인덱스와 함께 요소를 반환
for k, food in enumerate(foods):
  print(k, food)
```
