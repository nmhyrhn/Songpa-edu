# 문자열 str

```python
# 한 줄 문자열
s = '문자열'

print(s, type(s))

# 여러 줄 문자열
s = """
내용1
내용2
"""

print(s, type(s))

# 개행문자 사용
s2 = '\n내용3\n내용4\n'
print(s2, type(s2))
```

```python
# 문자열 연산
print("가나다" + "abc")
# print("가나다" - "abc") 불가함
print("가" * 50)

# print("가나다" + 123) # 불가함
print("가다나" + str(123)) 
print("가나다" + "123") 

#print("123" - 100)
print(int("123") - 100) 
```

# indexing | slicing

시퀀스 자료형에 대해 지원하는 기능

```python
# indexing
text = "hello world"
print(text[0]) # h
print(text[10]) # d
# print(text[11]) # error

print(len(text)) # 길이 str.length() 와 동일

# 역순 [-n]
print(text[-1]) # d
print(text[-11]) # h
```

```python
# sclicing
# [start:end:step] start번지부터 end번지전까지 반환

text = "hello world"
print(text[1:4])
print(text[6:9])
print(text[:4]) # 기본값: 0
print(text[4:]) # 기본값: len()
print(text[:])
print(text[6::1]) 
print(text[6::2])
print(text[::-1]) # 역순
print(text[6::-1]) # 역순
```

```python
alpha = 'abcdefg'
print(alpha[::2]) #aceg
print(alpha[1::2]) #bdf
print(alpha[::-1]) #gfedcba
```

# str api
str 자료형이 제공하는 함수(메소드)

```python
# 문자열 길이 확인
print(len('송파 여성인력 개발센터'))

# 포함여부 in 연산자
print('AI' in 'AI Agent')
print('ai' in 'AI Agent') # 대소문자 구분함

# 동등비교
print("AI" == "AI")
```

```python
# str.replace() : 문자열 치환
enroll_date = '2026/07/13'

print(enroll_date.replace('/', '-')) # out-of-place 연산
print(enroll_date) # 원본은 유지됨
```

```python
# str.strip() : 문자열 양쪽 공백 제거

data = '     안녕하세요       반갑습니다.      '
print(data)
print(data.strip())

# 대소문자 관련
original = "hELLO wORLD! 123"
print(original.upper()) # 대문자
print(original.lower()) # 소문자
print(original.capitalize()) # 첫글자만 대문자
print(original.swapcase()) # 대문자는 소문자로, 소문자는 대문자로
print(original.title()) # 각 단어의 첫글자만 대문자
```
