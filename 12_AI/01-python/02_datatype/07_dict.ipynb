# dict 사전

key, value로 요소를 구성해 보관하는 컨테이너 자료형.key를 통해 value를 찾는 방식이므로,
인덱싱/슬라이싱 지원하지 않는다.

- key: 중복을 허용하지 않음, immutable타입 값만 허용(int,tuple, str)
- value

```python
# dict(mutable) : 요소를 추가/수정/삭제 가능하다.

dict1 = {'a': 10, 'b': 20, 'c': 30}
print(dict1, type(dict1))

dict2 = {}
dict2['가'] = 'ga'
dict2['나'] = 'na'
print(dict2, type(dict2))
```

```python
# 값 조회
print(dict1['a'])
print(dict2['가'])
# print(dict1['xxxx']) # KetError; 'xxxx'
```

```python
# 중복을 허용하지 않음
dict1['a'] = 100

print(dict1)

# 요소 꺼내기
value = dict1.pop('a')
print(value)

print(dict1)

dict1['a'] = 100

print(dict1)
```

```python
# 다양한 타입의 value

scores = {
  'hong': [100, 90],
  'sinsa': [98, 90]
}

print(scores['hong'][1])
print(scores['sinsa'][0])
```

```python
user = {
  "name": "이순신",
  "age": 28,
  "hobby": ["넷플릭스", "수영"],
  "pet": {"name" : "댕댕이", "breed": "푸들"}
}

print(f"{user['name']}의 반려견의 이름은 {user['pet']['name']}입니다.")
```

```python
def add(a, b):
  return a + b

def multiply(a, b):
  return a * b

calculator = {
  "add" : add,
  "multiply" : multiply
}

print(add(10, 20))
print(calculator["add"](10, 20))
```

# dict 생성

```python
a = {}
a = dict() # 상동

print(a, type(a))

b = {'one': 1, 'two': 2}
b = dict(one=1, two=2) # 키워드인자 전달방식
print(b)

c = dict([('one', 1), ('two', 2)])
print(c)
```

# dict api

```python
# dict.get(key, default)

foods = {
  'donut': '🍩',
  'cake': '🎂',
  'candy': '🍬'
}

print(foods['donut'])
# print(foods['coffee'])

print(foods.get('donut'))
print(foods.get('coffee')) # None
print(foods.get('coffee', '😊')) # 존재 하지 않으면 기본값으로 넣어줘 😊
```

```python
# 반복순회
# dict.keys(): key만 반환(iterable)
# dict.values(): value만 반환(iterable)
# dict.items(): (key, value)묶음 반환 (iterable)

for key in foods:
  print(f'{key} = {foods[key]}')

print(foods.keys())
print(foods.values())
print(foods.items())
```

```python
for key in foods.keys():
  print(key)
for key in foods.values():
  print(key)

for item in foods.items():
  k, v = item
```

```python
# zip()
# 두 개 이상의 리스트의 요소를 묶어서 처리
names = ['홍길동', '신사임당', '이순신']
addresses = ['서울 관악구', '경기도 평택', '강원도 속초']
ages = (33, 28, 58)

for name, addr, age in zip(names, addresses, ages):
  print(f'이름: {name}, 주소: {addr}, 나이: {age}')
```

```python
# *list: list요소를 unpacking처리
data = [
  ('홍길동', '서울 관악구', 33),
  ('신사임당', '경기도 평택', 28),
  ('이순신', '강원도 속초', 56)
]

_names, _addrs, _ages = zip(*data)
print(list(_names))
print(list(_addrs))
print(list(_ages))
```
