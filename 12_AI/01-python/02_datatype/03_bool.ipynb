# 논리형 bool

- 참 True(1)
- 거짓 False(0)

```python
bool1 = True
bool2 = False

print(bool1, type(bool1))
print(bool2, type(bool2))
```

```python
print(1 == 3)
print(1 != 3)
print(1 != 1)
print()
print(10 > 4) #gt : greater than
print(10 >= 4) #gte : greater than or equal to
print(10 < 4) #lt : less than
print(10 <= 4) #lte : less than or equal to
```

```python
# 반전
bool3 = True
# print(!bool3) # False 불가능함!
print(not bool3) # False
```

```python
# 논리비교 연산자
# A and B : A가 참이면서, B도 참인가?
print(True and True) # True
print(True and False) # False
print(False and True) # False
print(False and False) # False
print('-' * 50)

# A or B : A 또는 B가 참인가?
print(True or True) # True
print(True or False) # True
print(False or True) # True
print(False or False) # False
```

```python
# and, or 나 수식이 많을경우는 ()를 잘사용하자

print(10 > 3 and 25 - 3 > 20)
print((10 > 3) and (25 - 3 > 20))
```
