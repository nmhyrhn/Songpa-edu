# 파이썬의 자료형(Data Type)
**기본 자료형**과 **컨테이너 자료형**으로 나눌 수 있다.

### 1. **기본 자료형 (Primitive Types)**

| 자료형 | 설명 | 예시 |
|--------|------|------|
| `int` | 정수형 | `10`, `-3`, `0` |
| `float` | 실수형 | `3.14`, `-0.1` |
| `bool` | 불리언 | `True`, `False` |
| `str` | 문자열 | `"hello"`, `'파이썬'` |
| `NoneType` | 값이 없음을 의미 | `None` |

---

### 2. **컨테이너 자료형 (Container Types)**

| 자료형 | 설명 | 예시 |
|--------|------|------|
| `list` | 순서 있고 변경 가능한 여러 값 | `[1, 2, 3]`, `["a", "b", "c"]` |
| `tuple` | 순서 있고 변경 불가능한 값 | `(1, 2, 3)` |
| `set` | 중복 없는 집합, 순서 없음 | `{1, 2, 3}` |
| `dict` | 키-값 쌍의 집합 | `{"name": "Tom", "age": 20}` |

---

### 3. **기타 자료형 (사용자 정의 포함)**

- `complex`: 복소수 (`1 + 2j`)
- 사용자 정의 클래스 객체 (`class`로 정의한 인스턴스)
- 함수, 모듈, 파일 등도 모두 객체로 간주됨



## 자료구조적 구분

### 1. **시퀀스형 (Sequence Type)**

- **순서가 있는 자료형**이다.
- 인덱싱, 슬라이싱이 가능하다.
- `len()`, `in`, 반복문 사용 가능하다.

**대표적인 시퀀스형**
| 자료형 | 예시 |
|--------|------|
| `list` | `[1, 2, 3]` |
| `tuple` | `(1, 2, 3)` |
| `str` | `'hello'` |
| `range` | `range(0, 10)` |

**특징**
- 인덱스 접근: `x[0]`
- 슬라이싱: `x[1:3]`
- 반복 가능: `for i in x`
- 길이: `len(x)`

---

### 2. **이터러블형 (Iterable Type)**

- **반복 가능한 객체**를 뜻한다.
- `for` 문으로 순회 가능
- `iter()` 함수로 **이터레이터(iterator)**를 만들 수 있음

**대표적인 이터러블형**
| 자료형 | 예시 |
|--------|------|
| `list`, `tuple`, `str`, `dict`, `set`, `range` | `for i in x:` |
| `file object` | `for line in open('file.txt'):` |
| 사용자 정의 iterable (`__iter__()` 구현) | class 기반 |

**이터러블과 이터레이터 차이**
| 구분 | 설명 |
|------|------|
| 이터러블 | 반복 가능한 객체 (ex: `list`, `str`) |
| 이터레이터 | `next()`로 값을 꺼낼 수 있는 객체 (`iter()`로 생성됨) |

```python
lst = [1, 2, 3]       # 이터러블
it = iter(lst)        # 이터레이터
print(next(it))       # 1
print(next(it))       # 2
```

---

### 정리

| 구분 | 시퀀스형 | 이터러블형 |
|------|----------|-------------|
| 순서 | 있음 | 있을 수도, 없을 수도 있음 |
| 인덱싱 | 가능 | 대부분 불가능 |
| 슬라이싱 | 가능 | 대부분 불가능 |
| 반복 | 가능 | 가능 |
| 예시 | `list`, `tuple`, `str`, `range` | `list`, `set`, `dict`, `file`, `generator` 등 |
