```python
a = 100
b = 123.456

print(a, type(a))
print(b, type(b))
```

```python
# 산술 연산
m = 10
n = 3

print(m + n)  # 덧셈
print(m - n)  # 뺄셈
print(m * n)  # 곱셈
print(m / n)  # 나눗셈

print(m // n) # 몫
print(m % n)  # 나머지

print(m ** 2) # 거듭제곱
print(m ** 3) # 거듭제곱
```

```python
# 반올림/올림/내림(버림)
n = 7 / 15

print(n)

#round 함수
print(round(n)) # 반올림
print(round(n, 1)) # 반올림
print(round(n, 2)) # 반올림


```

```python
# 외부 math 모듈(.py)에서 제공되는 함수 사용
import math

print(math.ceil(n))   # 올림
# print(math.ceil(n), 1)   # 불가능함
print(math.ceil(n * 10) / 10) # 0.466 -> 4.66 -> 5.0 -> 0.5

print(math.floor(3.9)) 
```
