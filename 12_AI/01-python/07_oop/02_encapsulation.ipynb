# 캡슐화 encapsulation

- 속성/메소드를 하나의 클래스안에 작성
- private 속성을 선언해서 외부의 직접 접근을 방지하고, public 메소드로써만 사용하도록 강제화
- 캡슐화 원칙을 잘 지켰다면, 주요 속성값을 외부에 노출하지 않는 은닉화 효과까지 얻게된다.

```python
# 캡술화하지 않은 은행계좌 예제

class Account:

  def __init__(self, name, balance):
    self.name = name
    self.balance  = balance

  def info(self):
    print(f'{self.name}님의 잔액은 현재 {self.balance}원 입니다.')

# 계좌 객체 생성
acc = Account('홍길동', 1_000_000)
acc.info()

# 입금처리
acc.balance = acc.balance + 500_000
acc.balance = acc.balance * 500_000
acc.balance = acc.balance + -500_000

acc.info()
```

```python
# 캡슐화된 은행 계좌 예제

class Account:

  def __init__(self, name, balance):
    # __속성명: private 속성 클래스 외부에서는 접근이 안됨
    self.__name = name
    self.__balance = balance

  def info(self):
    print(f'{self.__name}님의 잔액은 현재 {self.__balance}원 입니다.')

  def deposit(self, money):
    """입금메소드"""
    # 유효성검사
    if money > 0:
      self.__balance += money
      print(f'정상적으로 입금처리 되었습니다. 현재잔액: {self.__balance:,}원')
    else:
      print('입금액이 유효하지 않습니다.')

  def withdraw(self, money):
    """출금메소드"""
    # 유효성검사
    if 0 > money <= self.__balance:
      self.__balance -= money
      print(f'정상적으로 출금처리 되었습니다. 현재잔액: {self.__balance:,}원')
    else:
      print('출금액이 유효하지 않습니다.')

# 계좌 객체 생성
acc = Account('홍길동', 1_000_000)
acc.info()

# acc.__balance = acc.__balance + 500_000 # Account' object has no attribute '__balance'
# 입금처리
acc.deposit(500_000)
acc.deposit(-500_000)
# 출금처리
acc.withdraw(500_000)
acc.info()

```

# getter/setter method

- getter: private 속성값을 반환하는 메소드
- setter: private 속성값을 수정하는 메소드

```python
class User:

    def __init__(self, name, point):
        self.__name = name # 이름(읽기/쓰기 가능)
        self.__point = point # 포인트(읽기 가능)
        self.__last_login = '90일전'  # 은닉화 관리

    @property # 속성처럼 접근 가능
    def point(self):
        """getter 메소드"""
        return self.__point
    
    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, name):
        self.__name = name

leess = User('이순신', 1000)
# print(leess.point())
print(leess.point)
print(leess.name)

leess.name = '리순신'
print(leess.name)
```
