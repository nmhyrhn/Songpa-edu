# 객체지향 프로그래밍 (object-oriented programming)

- 객체 (메모리조각)를 사용해서 프로그램을 전개하는 방식
- 클래스를 작성해서 여러개의 객체를 할당 할 수 있다.
- 속성(변수:객체가 알고있는것)/메소드(함수:객체가 할수 있는것)로 클래스를 구성해서 작성

## 클래스와 객체

```python
# 클래스(객체의 설계도) 작성
# - 관례적으로 클래스명은 재문자로 시작(pascal-casing)

class Person:

  def __init__(self, name, age):
    """
    생성자 메소드(객체가 생성될때, 객체의 속성 초기화를 담당)
    모든 클래스 하위메소드는 첫번째 매개변수로 self(현재 객체)를 가진다.
    """
    self.name = name
    self.age = age
  
  def hello(self, greeting='안녕'):
    print(f'{greeting}, 나는 {self.age}살 {self.name}이야~')

# 객체 생성
hong = Person('홍길동', 33)    
print(hong, type(hong))
print(hong.name, hong.age)
hong.hello()
hong.hello('여보세요')

sinsa = Person('신사임당', 48)
print(sinsa, type(sinsa))
```

# 클래스/인스턴스 속성/메소드

- 클래스 속성: 클래스에 저장되는 1개의 속성(인스턴스 공유)
- 인스턴스 (객체)속성: 각 인스턴스마다 다르게 저장되는 속성
- 클래스 메소드: 클래스별로 수행되는 메소드
- 인스턴스 메소드: 각 인스턴스별로 수행되는 메소드(인스턴스 속성을 사용)

```python
class SongpaStudent:
  # 클래스 속성
  COURSE_NAME = 'JAVA&Spring Backend developer AI AGENT Camp'

  def __init__(self, name, nth):
    # 인스턴스 속성
    self.name = name
    self.nth = nth
  
  def introduce(self):
    # 인스턴스 메소드
    print(f'안녕하세요, {self.nth}기 {self.name}입니다.')

  @classmethod # 데코레이터
  def goodmorning(cls):
    # 클래스객체를 첫번째 인자로 사용
    print(f'안녕하세요~ {cls.COURSE_NAME} 여러분. 입실체크하세요~')  

  @staticmethod
  def goodbye(name):
    # 클래스객체를 인자로 받지않는다.
    print(f'수고하셨습니다. {SongpaStudent.COURSE_NAME} {name}님, 퇴실체크하세요~')

student1 = SongpaStudent('홍길동', 1)
print(student1)

print(student1.name, student1.nth, student1.COURSE_NAME, SongpaStudent.COURSE_NAME)

# 인스턴스 메소드 호출
student1.introduce()

SongpaStudent.goodmorning()
SongpaStudent.goodbye('홍길동')
```

# 객체지향 설계 원칙

1. 캡슐화: 속성/메소드를 하나의 클래스안에 함께 선언하고, 속성의 외부접근을 막고, 메소드로써 처리하는 것
2. 상속 부모클래스의 속성/메소드를 자식클래스에서 선언없이 사용하는 것
3. 다형성: 상속을 기반으로 하나의 객체가 여러타입처럼 작동하는 것
4. 추상화: 클래스 설계 시 객체가 가져가야할 보편적인 속성/메소드를 추출
