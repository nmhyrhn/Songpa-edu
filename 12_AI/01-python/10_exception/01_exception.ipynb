# Exception | Error

**일반적인 구분**
- Exception: 정상적인 흐름으로 복구가능한 미약한 오류.(try-except구문으로 복구)
- Error: 복구불가한 심각한 오류(재시작, 소스코드 수정)
> Python에서는 Exception과 Error를 엄격하게 구분하지 않는다.

**1. 모든 것이 Exception이다**
```python
# 이름에 "Error"가 붙어도 모두 Exception의 하위 클래스다
TypeError        # Exception 하위
ValueError       # Exception 하위  
MemoryError      # Exception 하위
ZeroDivisionError # Exception 하위
```

**2. 대부분 try-except로 잡을 수 있다**
```python
# MemoryError도 잡을 수 있다 (Java의 Error와 다름)
try:
    big_list = [0] * (10 ** 10)  # 메모리 부족
except MemoryError:
    print("메모리 부족 - 다른 방법 시도")
```

**3. 예외 계층 구조**

```

    ├── TypeError
```

## try..except

(중요) 발생한 예외/오류 상황에서 try..except구문을 통해 정상흐름을 돌려내야한다.

````python
try:
    print(10/0)
except Exception as e: # 발생한 예외객체
    # 예외/오류가 발생했을 떄 실행할 코드 작성
    print(e)

print('프로그램 정상 운영중')
````

````python
# 여러 타입의 예외처리
import random

try:
    n = random.randint(0, 2) # start~stop(포함) 사이의 임의의 정수 반환

    if n == 0:
        print(10 / 0)
    elif n == 1:
        int('ㅋㅋㅋㅋㅋ')
    else:
        lst = [1, 2, 3]
        print(lst[100])
except ZeroDivisionError as e:
    print('0으로 나누시면 안됩니다.')
except IndexError as e:
    print('존재하지 않는 인덱스를 참조하시면 안됩니다.')
except Exception as e:
    print('오류가 발생했습니다 : ', e)

print('프로그램 정상 운영중....')
````

````python
# try 블럭
# except블럭: 예외발생 시 실행되는 블럭
# else블럭(선택) : try블럭이 예외없이 완료된 경우 실행되는 블럭
# finally블럭(선택) : 모든 경우에 실행되는 블럭

import traceback

try:
    f = open('hello2.txt', 'r', encoding='utf-8')
except Exception as e:
    print(e)
    traceback.print_exc()
else:
    # 정상흐름에서 처리구문
    print('내용: ', f.read())
    f.close()
finally:
    print('무조건 실행')

````

````python
# 예외처리 pass(예외무시)는 디버깅을 어렵게해서 좋지 않음
try:
    print(10 / 0)
except:
    pass # 확실한 경우에만 사용

print('-' * 5)
````
