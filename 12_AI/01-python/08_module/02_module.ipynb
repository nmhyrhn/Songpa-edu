# 모듈 module

- .py 파일을 의미
- 코드 재사용성을 높이기 위해 모듈단위로 관련있는 것들을 묶어서 관리
- 모듈하위의 변수, 함수, 클래스 등을 외부에서 임포트 후 사용이 가능
- private 변수 선언시에는 외부에서 접근 가능

```python
import math

print(dir(math)) # 특정모듈의 사용가능한 변수들
print(dir()) # 현재 모듈의 사용 가능한 변수
```

```python
# 현대 모듈이름
print(__name__)
# import한 모듈 이름
print(math.__name__)
```

```python
# 모듈 import를 추천
from songpawork import my_math

print(my_math.x)
print(my_math.pi)
print(my_math.get_circle_area(10))
```

```python
# 모듈 import를 추천
from songpawork import my_math as m

print(m.x)
print(m.pi)
print(m.get_circle_area(10))
```

```python
# 현재 모듈의 변수와 이름충돌이 우려
from songpawork.my_math import x, pi, get_circle_area
# from songpawork.my_math import * # 비추

print(x)
print(pi)
# print(__z) *에서는 오류 발생
print(get_circle_area(10))
```
