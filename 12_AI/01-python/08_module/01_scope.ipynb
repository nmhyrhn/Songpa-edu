# 유효범위 scope

- 변수에 접근 가능한 범위
- global scope: 전역레벨에 작성한 변수. 어디서나 접근이 가능하다.
- local scope: 블럭하위에 선언된 변수. 블럭을 벗어나면 접근이 불가능하다.

```python
# 전역변수
a = 100
print(a)

def foo():
  print(a)

foo()
```

```python
# 노트북의 다른 코드셀에서도 전역벽수는 접근 가능하다.
print(a)
```

```python
# 변수 우선순위: 가까운 변수를 먼저 사용
x = 33 # 전역변수

def koo():
  x = 55 # 지역변수
  print(x)

koo()
print(x)
```

# call by value | call by reference

- call by value 값으로 함수호출(immutable)
- call by reference 참조주소값으로 함수호출(mutable)

```python
# call by value
def f1(v):
  v = 200
  return v

v = 100
# f1(v)
v = f1(v)
print(v)
```

```python
# call by reference
def f2(lst):
  lst[1] *= 100

lst = [1, 2, 3]
print(lst, type(lst))
# dir(lst)
f2(lst)
print(lst)
```
