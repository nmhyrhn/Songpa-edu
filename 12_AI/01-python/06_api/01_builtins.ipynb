# map

- iterable 객체의 각 요소에 대하여 지정한 함수를 적용하고 난 결과를 map 객체로 반환

```python
texts = ['토마토 쥬스', '까페아메리카노', '코코아라떼', '식혜']

# 리스트내포
lens = [len(text) for text in texts]

print(lens)

# map
# lens = [*map(len, texts)]
lens = [*map(lambda x: len(x), texts)]
lens
```

```python
# 여러 iterable 객체 통합해서 사용
mid_terms = [90, 70, 66]
final_terms = [100, 80, 60]

list(map(lambda mid, final: (mid + final) / 2, mid_terms, final_terms))
```

# filter

주어진 iterable객체에 대해 지정한 함수를 적용 후 Ture값만 최종 리스트 객체에 포함

```python
nums = [n + 1 for n in range(10)]

nums

# 리스트내포
events = [n for n in nums if n % 2 == 0]

events

# filter 함수 적용
evens = list(filter(lambda x: x % 2 == 0, nums))
evens
```

```python
names = ['홍길동', '고길동', '최홍만', '홍진경', '최길동']
# str.startswith('홍')
print(names[0].startswith('홍'))

hongs = [name for name in names if name.startswith('홍')]
hongs
```

# reduce

functools 모듈에서 제공. iterable 객체의 요소에 대하여 지정한 연산 후 단일값을 반환하는 함수

**매개변수**

1. 함수(누적연산값, 현재값) -> 결과값
2. iterable 객체
3. 초기값(선택): 생략 시 첫번째 인자가 초기값으로 사용

```python
from functools import reduce

# 1~10까지의 합
nums = [n for n in range(1, 11)]

total = reduce(lambda acc, current: acc + current, nums, 0)
print(total)
```

```python
def f(acc, current):
  """연산과정을 보기위한 함수"""
  print(acc, current)
  return acc + current

total = reduce(f, nums, 0)
print(total)
```

```python
# 최대값 조회하기
numbers = [23, 5, 6, 100, 399, 342, -102, 435, 12]

max_num = reduce(lambda a, b: a if a > b else b, numbers)
max_num
```

```python
def op(a, b):
  print(f'{a= }, {b= } : max = {a if a> b else b}')
  return a if a > b else b

max_num = reduce(op, numbers)
max_num
```
