# 날짜/시각 datetime

## 날짜/시각

```python
from datetime import datetime

today = datetime.today()
print(today, type(today))

now = datetime.now()
print(now, type(now))
```

```python
print(now.year)
print(now.month)
print(now.day)
print(now.hour)
print(now.minute)
print(now.second)

print(now.date())
print(now.time())
```

```python
weekdays=['월', '화', '수', '목', '금', '토', '일']

weekdays[now.weekday()]
```

### 형식표현
https://docs.python.org/3/library/datetime.html#strftime-and-strptime-format-codes

- `%Y`: 연도 (예: 2024)
- `%m`: 월 (01부터 12)
- `%d`: 일
- `%H`: 시간 (24시간 형식)
- `%M`: 분
- `%S`: 초

![형식문자표](https://d.pr/i/v8vL7u+)

```python
# 날짜/시간 -> 문자열
print(now.strftime('%Y-%m-%d %H:%M:%S'))
print(now.strftime('%y-%m-%d %I:%M:%S'))
print(now.strftime('%y-%m-%d (%a) %I:%M:%S'))
```

```python
# 문자열 -> 날짜/시간
text = '26-07-15 09:31:38'
parsed_date = datetime.strptime(text, '%y-%m-%d %I:%M:%S')
parsed_date
print(parsed_date.date())
```

```python
# d-day 계산
from datetime import date

# today = datetime.today().date()
today = date.today()
target_date = date(2026, 12, 24)

diff = target_date - today
dday = diff.days
print(f'크리스마스까지 D-{dday}일 남았습니다.')
```
